### Analyzing Subsets of Data with .iloc[ ]

[.iloc[ row criteria , colun criteria ]](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html) ... 

The iloc property gets, or sets, the value(s) of both row and column of the specified indexes, e.g. dataframe_name.iloc[row, column]

To access more than one row, use double brackets and specify the indexes, separated by commas:

--> df.iloc[[0, 2]]

Specify columns by including their indexes in another list:

--> df.iloc[[0, 2], [0, 1]]

You can also specify a slice of the DataFrame with from and to indexes, separated by a colon:

--> df.iloc[0:2]

**Note**: When slicing, the to index is excluded from the result.

Return Value Depends on the input:

- Single indexes for both row and column [1, 0] returns the content of that cell.

- Single index for one row [1] returns a Pandas Series.

- A list of indexes [[0, 2]] returns a Pandas DataFrame.

### Analyzing Subsets of Data
page 10 - 11

[.loc[ row criteria , colun criteria ]](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html) ... Access a group of rows and columns by filtering the data using label(s) or a boolean array. This lets you insert ranges and boolean expressions for the rows along with column names as strings or a list of column names for the columns. The .loc[ ... , ... ] can also have methods chained as well as be part of compound logical expressions. 

```python
# How many total rides were between 3:00am and 4:00am on January 1
# Thus, wow would I find the value for the fourth hour's count (remember, zero indexing)
hourly_data.loc[3,'count']
```

```python
# What if I wanted to see the number of casual, registered and total rides per hour between 2:00am - 5:00am on January 1
# Note, the slice 2:4 is inclusive of rows 2, 3, & 4... 
hourly_data.loc[2:4, ['casual','registered', 'count']]
```

```python
# For the year, for every hour from midnight to 5:00am (but not including the 5am hour), what is the average casual, registerd and total count values
# here we can use a boolean condition to subset the rows
hourly_data.loc[hourly_data['hr'] < 5, ['casual','registered','count']].mean()
```

```python
# how does 12am - 4am ridership vary between cold and hot nights? Let's look at cold nights, temp <= 0.50
# here we use a logical statement of (boolean condition) & (boolean condition) to subet the rows <-- note parenthesis ( ) around conditions with & for and
hourly_data.loc[(hourly_data['hr'] < 5) & (hourly_data['temp'] <= 0.50) , 'count'].mean() 
```

```python
# how does 12am - 4am ridership vary between cold and hot nights? Let's look at hot nights, temp > 0.50
# here we use a logical statement of (boolean condition) & (boolean condition) to subet the rows <-- note parenthesis ( ) around conditions
hourly_data.loc[(hourly_data['hr'] < 5) & (hourly_data['temp'] > 0.50) , 'count'].mean() 
```

```python
# how does early AM ridership vary with it is hot OR humid? Let's look at hot or humid nights, temp > 0.50 | hum > 0.50
# here we use a logical statement of (boolean condition) | (boolean condition) to subet the rows <-- note parenthesis ( ) around conditions with pipe | for or
hourly_data.loc[(hourly_data['hr'] < 5) & ((hourly_data['temp'] > 0.5) | (hourly_data['hum'] > 0.50)) , 'count'].mean() 
```

```python
# seeing that the mean for both warm and humid nights is lower than just warm nights it must indicate that:
# 1) there are a lot of humid nights and
# 2) that ridership is down for those hot and humid nights
# So lets see the average rides per hour when it is humid
hourly_data.loc[(hourly_data['hr'] < 5) & (hourly_data['hum'] > 0.50) , 'count'].mean() 
```

```python
# we can count the number of humid hours between mindight and 5:00am
hourly_data.loc[(hourly_data['hr'] < 5) & (hourly_data['hum'] > 0.50) , 'count'].count()
```

```python
# and compare that to the number of hot overnight hours
# we can see that with 2x the humber of humid hours that people indeed ride less when humid vs hot
hourly_data.loc[(hourly_data['hr'] < 5) & (hourly_data['temp'] > 0.50) , 'count'].count()
```

[.groupby()](https://pandas.pydata.org/docs/user_guide/groupby.html) By “group by” we are referring to a process involving one or more of the following steps:
- Splitting the data into groups based on some criteria.
- Applying a function to each group independently.
- Combining the results into a data structure.

```python
# for the data frame hourly_data
# Split: use the groupby() method to group by each unique value within ['column name to be grouped by']
# Apply: once grouped, take the .mean() of the 'column name you want to analyze by group'
# here the question is: how many rides per day did we average per each of the four seasons?
# the code shows us the average = mean(), rides = ['count'], grouped by season ['season'] 
hourly_data.groupby(['season'])['count'].mean()
```

```python
# you can also subset your subset like a pivot table
# here we want to see the rides by season then subset between workdays and non-workdays
hourly_data.groupby(['season', 'workingday'])['count'].mean()
```

```python
# importing matplotlib.pyplot as plt is convention
import matplotlib.pyplot as plt

# create .subplots() along with a figure size (width, height) tuple
fig, ax = plt.subplots(figsize = (20, 6))

# use .scatter() to assign the features to each axis
ax.scatter( x = hourly_data['instant'], y = hourly_data['count'])

# label each axis and give the chart a title
plt.xlabel('Hour')
plt.ylabel('# of rides')
plt.title("Number of rides per hour")

plt.show()
```

```python
# let's look at the first ten days to see the data better
# use .loc() to slice a cut of the complete data set to see only the first 240 hours (10 days)
first_ten_days = hourly_data.loc[0:240, :]

# create the subplot
fig, ax = plt.subplots(figsize = (20, 6))

# use .scatter() to identify the x and y features, along with the marker color and symbol
ax.scatter(x = first_ten_days['instant'], y = first_ten_days['count'], c = 'red', marker = '|')

# label each axis and title the chart
plt.xlabel("Hour")
plt.ylabel("# of Rides")
plt.title("Rides per hour for the first ten days")

# display the visual
plt.show()
```

### Using .plot() versus .scatter()
page 17

[matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
- The .scatter() will just plot the points based on the x , y inputs
- The .plot() function is used to draw points (markers) in a diagram. By default, the plot() function draws a line from point to point (but can be changed).
- The function takes parameters for specifying points in the diagram.
- Parameter 1 is an array containing the points on the x-axis.
- Parameter 2 is an array containing the points on the y-axis.

```python
fig, ax = plt.subplots(figsize = (20, 6))
ax.plot(first_ten_days['instant'], first_ten_days['casual'], c = 'red', label = 'non-member', linestyle = '-')
ax.plot(first_ten_days['instant'], first_ten_days['registered'], c = 'blue', label = 'member', linestyle = '--')
ax.legend()
plt.show()
```

- The whiskers extend from the box to the farthest data point lying within 1.5x the inter-quartile range (IQR) from the box. 
- Flier points are those past the end of the whiskers.

Let's say, for our registered members, we want to see how many rides happen per hour (page 18)

```python
import seaborn as sns
fig, ax = plt.subplots(figsize = (20, 6))
sns.boxplot(x = 'hr', y = 'registered', data = hourly_data)
plt.xlabel("Hour (on a 24-hour clock)")
plt.ylabel("# of rides")
plt.title("Registered Member Rides per hour")
plt.show()
```

```python
# same but for 'casual' non-member riders
fig, ax = plt.subplots(figsize = (20, 6))
sns.boxplot(x = 'hr', y = 'casual', data = hourly_data)
plt.xlabel("Hour (on a 24-hour clock)")
plt.ylabel("# of rides per hour")
plt.title("Non-Member rides per hour")
plt.show()
```
